```python
df=pd.read_csv('D:\\Drive into Analysis\\IPL Data Analysis\\1731674068078_TATA IPL 2025- Auction List -15.11.24.csv')
df.head(5)
```

```python
df.info()
```

```python
df.drop(columns=['Unnamed: 6','Unnamed: 18'],inplace=True)
df.rename(columns={'Unnamed: 10':'Batting Style','Unnamed: 11':'Bowling Style','Unnamed: 17':'Current_Team'},inplace='True')
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df['DOB']=pd.to_datetime(df['DOB'],format="%d/%m/%Y")
```

```python
df.dtypes
```

```python
df.duplicated().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
labels=['RHB','LHB']
plt.bar(labels,df['Batting Style'].value_counts(),color='blue')
plt.xlabel('Type of batsman')
plt.ylabel('Count')
plt.title('Total count of batman')
```

```python
df.select_dtypes(include=['int64','float64']).corr()
```

```python
sns.regplot(x=df['IPL'],y=df['Price Rs'],color='b')
```

```python
df.isna().sum()
```

```python
df['Test caps'].fillna(0,inplace=True)
df['Bowling Style'].fillna('None',inplace=True)
df['ODI caps'].fillna(0,inplace=True)
df['T20 caps'].fillna(0,inplace=True)
df['Current_Team'].fillna('None',inplace=True)
df['Previous IPLTeam(s)'].fillna('None',inplace=True)
df['C/U/A'].fillna('None',inplace=True)
df['IPL'].fillna('None',inplace=True)
avg=df['Price Rs'].value_counts().idxmax()
df['Price Rs'].fillna(avg,inplace=True)
```
